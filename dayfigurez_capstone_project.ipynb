```python
# running hyper perameter tunning to see if we can get the best model
#
mlflow.set_tracking_uri('./mlruns')
mlflow.set_experiment("dayfigurez_capstone_project")

# Define the models and their hyperparameter grids
models = {
    "Logistic_Regression": {
        "model": LogisticRegression(random_state=42, max_iter=1000,class_weight='balanced'),
        "params": {
                'C': [0.01, 0.1, 1, 10, 100],
                'solver':['liblinear','saga']
                }
    },
    "KNN": {
        "model": KNeighborsClassifier(),
        "params": {
            'n_neighbors': [3, 5, 7],
            'weights': ['uniform', 'distance']
            }
    },
    "Random_Forest": {
        "model": RandomForestClassifier(class_weight='balanced'),
        "params": {
            'n_estimators': [100, 150, 200],
            'max_depth': [None, 10, 20],
            'min_samples_split':[2,5,10]
            # 'min_leaf':[1,2,4]
            }
    },
    'Decision_Tree':{
        'model': DecisionTreeClassifier(random_state=42,class_weight='balanced'),
        'params':{
            'criterion':['gini','entropy'],
            'max_depth':[None,10,20,30],
            'min_samples_split':[2,5,10]
            }
        },
        "XGBoost": {
        "model": XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss'),
        "params": {
            'n_estimators': [50, 100],
            'max_depth': [3, 5, 7],
            'learning_rate': [0.05, 0.1],
            'subsample': [0.8, 1.0],
            'colsample_bytree': [0.8, 1.0],
            'gamma': [0, 1]
        }
    }
}


# Variables to keep track of the best overall model
best_name = ""
best_model = None
best_F1_score = 0.0
best_auc =0.0
best_recall_score =0.0


for model_name, config in models.items():
    with mlflow.start_run(run_name=f"Tuning_{model_name}"):
        print(f"TRAINING AND TURNING {model_name}...\n"+'~'*30)

        # Initialize GridSearchCV
        grid_search = GridSearchCV(
            estimator=config["model"],
            param_grid=config["params"],
            cv=5,
            n_jobs=-1,
            scoring='accuracy'
        )

        # Fit the model
        grid_search.fit(x_train_scale_sample, y_train_sample)

        # Extract best model and parameters for this specific models_group
        best_model_frm_group = grid_search.best_estimator_
        best_params_frm_group = grid_search.best_params_

        # Evaluate on the test set
        y_pred = best_model_frm_group.predict(x_test_scale_sample)
        test_accuracy = accuracy_score(y_test_sample, y_pred)
        test_precision = precision_score(y_test_sample, y_pred)
        test_F1_score =f1_score(y_test_sample,y_pred)
        test_auc_score =roc_auc_score(y_test_sample,y_pred)
        test_recall_score =recall_score(y_test_sample,y_pred)

        # Log to MLflow
        mlflow.log_params(best_params_frm_group)
        mlflow.log_metric("test_accuracy", test_accuracy)
        mlflow.log_metric("roc_auc_score", test_auc_score)
        mlflow.sklearn.log_model(best_model_frm_group, f"{model_name}_model") # type: ignore

        print(f"[{model_name}] Best Params>>: {best_params_frm_group}")
        print(f"[{model_name}] Test Accuracy>>: {test_accuracy * 100:.2f}%")
        print(f'[{model_name}] test auc score>>: {test_auc_score* 100:.2f}%')
        print(f'[{model_name}] test f1 score>>: {test_F1_score*100:.2f}%')
        print(f'[{model_name}] test recall score>>: {test_recall_score*100:.2f}%')
        print(f"[{model_name}] Test precision>>: {test_precision * 100:.2f}%\n")

        # Check if this is the best model overall
        if test_auc_score > best_auc and test_F1_score>best_F1_score and test_recall_score>best_recall_score:
            best_F1_score = test_F1_score
            best_model = best_model_frm_group
            best_name = model_name
            best_recall_score =test_recall_score


print("~" * 40)
print(f"🏆 BEST MODEL: {best_name} with {best_F1_score * 100:.2f}%  and {best_recall_score * 100:.2f%} and {best_auc:.4f}")
```
